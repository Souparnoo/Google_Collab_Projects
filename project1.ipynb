```python
import requests,os

# Download a sample image
url= "https://hips.hearstapps.com/clv.h-cdn.co/assets/16/18/gettyimages-586890581.jpg?crop=0.668xw:1.00xh;0.219xw,0"
r = requests.get(url)
with open("test.jpg", "wb") as f:
    f.write(r.content)
print(os.path.exists('test.jpg'))   # should print True
print(os.path.getsize('test.jpg'))  # should print a number > 0
```

```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('test.jpg')

if img is None:
    print("ERROR: image not loaded — check the file exists")
else:
    print(f"Success! Image shape: {img.shape}")
    cv2_imshow(img)
```
